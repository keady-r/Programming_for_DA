```python
breastCancer = pd.read_csv("DataSets/BreastCancerDataSet.csv")
```

```python
breastCancer_subsection = breastCancer.loc[1:,["diagnosis","radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean"]]
```

```python
malignant = breastCancer.loc[breastCancer["diagnosis"]=="M"] 
benign = breastCancer.loc[breastCancer["diagnosis"]=="B"]
```

```python
plt.subplot(1, 2, 1)
sns.scatterplot(data = malignant, x = "area_mean", y = "perimeter_mean", legend = True, color = 'red')
sns.scatterplot(data = benign, x= "area_mean", y = "perimeter_mean", legend = True, color = 'purple')
plt.legend(labels=["Malignant", "Benign "])
plt.title("Perimeter vs Area")

plt.subplot(1, 2, 2)
sns.scatterplot(data = malignant, x = "symmetry_mean", y = "smoothness_mean", legend = True, color = 'red')
sns.scatterplot(data = benign, x= "symmetry_mean", y = "smoothness_mean", legend = True, color = 'purple')
plt.title("Symmetry vs Smoothness")
plt.legend(labels=["Malignant", "Benign "])

plt.show()
plt.clf()
```

```python
plt.subplot(1, 2, 1)
plt.hist( malignant['area_mean'], color = 'red', edgecolor = 'black',
         bins = int(180/5))
#Labels
plt.title('Hist: mean area of Malignant Case', size = 9)
plt.xlabel('area_mean', size = 9)
plt.ylabel('Number of cases', size = 9)

plt.subplot(1, 2, 2)
plt.hist( benign['area_mean'], color = 'purple', edgecolor = 'black',
         bins = int(180/5))
#labels
plt.title('Hist: mean area of Benign Case', size = 9)
plt.xlabel('area_mean', size = 9)
plt.ylabel('Number of cases', size = 9)
```

```python
plt.subplot(1, 2, 1)
plt.hist( malignant['concavity_mean'], color = 'red', edgecolor = 'black',
         bins = int(180/5))
#Labels
plt.title('Hist: mean concavity of Malignant Case', size = 9)
plt.xlabel('concavity_mean')
plt.ylabel('Number of cases')

plt.subplot(1, 2, 2)
plt.hist( benign['concavity_mean'], color = 'purple', edgecolor = 'black',
         bins = int(180/5))
#labels
plt.title('Hist: mean concavity of Benign Case', size = 9)
plt.xlabel('concavity_mean')
plt.ylabel('Number of cases')

```
